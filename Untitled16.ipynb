```python
#1111
import cv2
import matplotlib.pyplot as plt
import numpy as np
# Dosyayı BGR formatında okuma
degerli_bgr = cv2.imread("para.png")
```

```python
#2222
import cv2
import matplotlib.pyplot as plt
import numpy as np
# BGR den RGB ye dönüştürme
degerli_rgb = cv2.cvtColor(degerli_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(degerli_rgb)
```

```python
#3333
import cv2

# RGB resmi gri tonlamaya çevir
degerli_gri = cv2.cvtColor(degerli_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(degerli_gri)
```

```python
#4444
import cv2

# Bulanıklaştırma
degerli_sansür = cv2.GaussianBlur(degerli_gri, (5, 5), 0)

plt.imshow(degerli_sansür)
```

```python
#5555
import cv2

# ikili resim elde eymme
ret, degerli_ikili = cv2.threshold(degerli_sansür, 180, 255, cv2.THRESH_BINARY)

plt.imshow(degerli_ikili)
```

```python
#6666
import cv2

# kenar bulma
degerli_kenarlar = cv2.Canny(degerli_ikili, 60, 200)

plt.imshow(degerli_kenarlar)
```

```python
#7777
import cv2
import numpy as np

konturlar, hierarchy = cv2.findContours(degerli_kenarlar, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

degerli_konturlar = degerli_rgb.copy()

for kontur in konturlar:
    if cv2.contourArea(kontur) > 180:
        x, y, w, h = cv2.boundingRect(kontur)
        cv2.rectangle(degerli_konturlar, (x, y), (x + w, y + h), (0, 255, 0), 2)

        (cx, cy), ycapi = cv2.minEnclosingCircle(kontur)
        cv2.circle(degerli_konturlar, (int(cx), int(cy)), int(ycapi), (255, 0, 0), 2)

        harfler = "Y.O."
        # Y harfi ortada gözüküyor ama Y.O. olunca ortada değilde sağdaymış gibi gözüküyor -15 olmasının nedeni bu
        cv2.putText(degerli_konturlar, harfler, (x + w//2 - 15, y + h//2),

                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

plt.imshow(degerli_konturlar)

```

```python
#8888
import cv2
import numpy as np

parabilgi = []
for kontur in konturlar:
    alan = cv2.contourArea(kontur)
    if alan > 200:
        parabilgi.append((kontur, alan))

# alanlarına göre sıralama
parabilgi.sort(key=lambda x: x[1])

degerli_numarali = degerli_rgb.copy()

for eva, (kontur, alan) in enumerate(parabilgi):
    # çizim
    x, y, w, h = cv2.boundingRect(kontur)
    cv2.rectangle(degerli_numarali, (x, y), (x + w, y + h), (0, 255, 0), 2)


    # sıralama
    paranumara = eva + 1
    cv2.putText(degerli_numarali, str(paranumara), (x + w%2, y + h%2),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

plt.imshow(degerli_numarali)
```
